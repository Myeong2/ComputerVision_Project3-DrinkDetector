```python
from google.colab import drive
drive.mount('/content/drive/')

import os
from PIL import Image

# 이미지 파일이 존재하는 폴더
input_folder = '/a. image/dog'
# 조정된 이미지를 저장할 폴더
output_folder = '/content/drive/MyDrive/output_folder2'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for file in os.listdir(input_folder):
    if file.endswith('.jpg') or file.endswith('.png'):    # 이미지 파일 찾기
        image_path = os.path.join(input_folder, file)
        image = Image.open(image_path)
        resized_image = image.resize((128, 128))    # 이미지 크기 조정

        # 조정된 이미지 저장
        output_path = os.path.join(output_folder, file)
        resized_image.save(output_path)
        print(f'이미지 {file} 처리 완료: 크기 변경 및 저장 완료')
```
